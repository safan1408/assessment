```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# pmf of poisson

```python
plt.gcf().set_size_inches(12,6)
for l in range(3,8,1):
    seq=np.arange(0,20)
    distribution = np.zeros(seq.shape[0])
    for i in range(seq.shape[0]):
        distribution[i] = stats.poisson.pmf(i,l)

    plt.plot(seq, distribution)
    plt.legend(["\u03BB=3","\u03BB=4",'\u03BB=5','\u03BB=6','\u03BB=7'])
    plt.ylabel('probability')
    plt.xlabel('Number of Events')
    plt.title("PDF-Varying \u03BB")
    
```

# cdf of poisson 

```python
plt.gcf().set_size_inches(12,6)
for l in range(3,8,1):
    seq=np.arange(0,20)
    distribution = np.zeros(seq.shape[0])
    for i in range(seq.shape[0]):
        distribution[i] = stats.poisson.cdf(i,l)

    plt.plot(seq, distribution)
    plt.legend(["\u03BB=3","\u03BB=4",'\u03BB=5','\u03BB=6','\u03BB=7'])
    plt.ylabel('Cumulative probability')
    plt.xlabel('Number of Events')
    plt.title("CDF-Varying \u03BB")
    
```
