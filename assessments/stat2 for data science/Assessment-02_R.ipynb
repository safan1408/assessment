```r
a=rchisq(n = 100000,df = 1)
b=rchisq(n=100000,df=2)
c=rchisq(n=100000,df=3)
d=rchisq(n=100000,df=6)
p=ggplot(mapping=aes(x=a))+geom_density(fill='steelblue2')+labs(title = "Chi Square distribution for df=1",x="x")
q=ggplot(mapping=aes(x=b))+geom_density(fill='steelblue2')+labs(title = "Chi Square distribution for df=2",x="x")
r=ggplot(mapping = aes(x=c))+geom_density(fill='steelblue2')+labs(title = "Chi Square distribution for df=3",x="x")
s=ggplot(mapping = aes(x=d))+geom_density(fill='steelblue2')+labs(title = "Chi Square distribution for df=6",x="x")
ggarrange(p, q, r, s, ncol = 2, nrow = 2)
```
